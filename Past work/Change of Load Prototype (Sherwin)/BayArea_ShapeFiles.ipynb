# Census Tracts files

This defines all California census tracts.

```python
census_tracts = '/Users/melek/Desktop/Research/Research Fall 2023/DER EJ/Data/tl_2024_06_tract/tl_2024_06_tract.shp'
CTs = gpd.read_file(census_tracts)
CTs = CTs.to_crs(epsg=4326)
display(CTs)
```

## Example

Below is an example of defining a point and using it to get a census tract. For example, let put in the coordinates of [EBU2](https://maps.app.goo.gl/kqMY5KZSKUrXiPoy8)

```python
latitude = 32.881178 # coordinate ending in North or South
longitude = -117.233350 # coordinate ending in East or West
point = Point(longitude, latitude)
census_tract = CTs[CTs.contains(point)]
census_tract['NAMELSAD']
```

```python
import os

# Define the path to your main folder containing the 25 folders
main_folder = '/Users/melek/Desktop/Research/Research Fall 2023/DER EJ/Data/out_gis/'

# Set to store unique names of .shp files
shapefile_names = set()

# Walk through each sub-folder inside the main folder
for root, dirs, files in os.walk(main_folder):
    # Look for files that end in .shp
    for file in files:
        if file.endswith('.shp'):
            shapefile_names.add(file)  # Add file name to the set

# Convert the set to a list (if needed) and print unique names
unique_shapefile_names = list(shapefile_names)
for name in unique_shapefile_names:
    print(name)
```

```python
for name in unique_shapefile_names:
    print(name)
    gpd.read_file(os.path.join(main_folder,'P1R/GIS/Network_nodes.shp'))
```

```python
for root, dirs, files in os.walk(main_folder):
    # Look for files that end in .shp
    for file in files:
        if file.endswith('.shp'):
            shapefile_names.add(file)  # Add file name to the set

# Loop through unique shapefile names and check if each exists
for name in shapefile_names:
    found = False  # Flag to indicate if the file was found
    for root, dirs, files in os.walk(main_folder):
        # Construct the path to the shapefile
        file_path = os.path.join(root, name)
        # Check if the file exists
        if os.path.isfile(file_path):
            print(f"Displaying data for {name}")
            # Display the data with all columns
            display(gpd.read_file(file_path))
            found = True
            break  # Stop searching once the file is found
    if not found:
        print(f"{name} not found in any folder.")
```
